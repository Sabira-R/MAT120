# **Derive the given function**








```python
#T1

from sympy import*
x= symbols('x')

f = log(x)
df = diff(f,x)
df2 = diff(f,x,2)
print(df)
print(df2)

f= lambdify(x,f)
df = lambdify(x,df)

import numpy
x_value = numpy.arange(1, 26)
print(x_value)
test_number = numpy.arange(1, 26)
a= df(test_number)
print(a)

import matplotlib.pyplot as plt
plt.plot(test_number , a)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Shape of the function')
plt.show()
```

```python
#T2

from sympy import*
x= symbols('x')

f = sin(x)
df = diff(f,x)
df2 = diff(f,x,2)
print(df)
print(df2)

f= lambdify(x,f)
df = lambdify(x,df)

import numpy
x_value = numpy.arange(-10, 10)   #(-10, 10, 0.01)
print(x_value)
test_number = numpy.arange(-10, 10)    #(-10, 10, 0.01)
a= df(test_number)
print(a)

import matplotlib.pyplot as plt
plt.plot(test_number , a)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Shape of the function')
plt.show()
```
