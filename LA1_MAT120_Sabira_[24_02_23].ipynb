```python
#TASK-01

from sympy import*
def find_maxima_minima (f, x):
    df = diff(f, x)
    df2 = diff(f, x, 2)
    extrema = solve(diff(f, x), x)

    if extrema:
        try:
            print(f"Critical points at x= {tuple(extrema)}")
            points = []
            for i in extrema:
                    points.append((i, f.subs(x, i).evalf()))

            d_max, d_min = {}, {}
            for i, j in points:
                if df2.subs(x, i).evalf() < 0:
                    print(f"Point ({i}, {j}) is Local Maxima")
                    d_max[i] = j
                elif df2.subs(x, i).evalf() > 0:
                    print(f"Point ({i}, {j}) is Local Minima")
                    d_min[i] = j
                else:
                    print(f"Point ({i}, {j}) is a Saddle Point")
            d_max[-5], d_max[6] = f.subs(x, -5).evalf(), f.subs(x, 6).evalf()
            d_min[-5], d_min[6] = f.subs(x, -5).evalf(), f.subs(x, 6).evalf()
            max_val = max(d_max.values())
            min_val = min(d_min.values())
            print('\n'.join([f"Point ({i}, {j}) is a Global Maxima" for i, j in d_max.items() if j == max_val]))
            print('\n'.join([f"Point ({i}, {j}) is a Global Minima" for i, j in d_min.items() if j == min_val]))
        except:
                print("It is a Imaginary number.")

    else:
        print("No Maxima or Minima at interval (-5, 6)")
    plot(f, (x, -5, 6))

x = symbols("x")
f = 9*(x**3) - 7*(x**2) + 3*x + 10
find_maxima_minima (f, x)
```

```python
#TASK03

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

a, b = -5, 6

coeffs = [1, 0, -1]
roots = np.roots(coeffs)
x_crit = [x for x in roots if a <= x <= b]

y_crit = []
for x in x_crit:
    y_crit.append(f(x))
y_endpoints = [f(a), f(b)]

y_min = min(y_crit + y_endpoints)
y_max = max(y_crit + y_endpoints)

print(f"Minimum value: {y_min}")
print(f"Maximum value: {y_max}")

x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin x")
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.ylim(-1.5, 1.5)
plt.show()
```
