```python
import sympy as sym, matplotlib.pyplot as plt  # importing modules
t, k = sym.symbols('t k')  # Declaring Symbols
y = sym.Function('y')(t)    # Declaring Functions
solutions = {k_val: sym.dsolve(sym.diff(y, t) + k_val*y, ics={y.subs(t, 0): 5}).rhs for k_val in [0.1, 0.2, 0.5]}  # Dictionary Comprehension
for k_val, sol in solutions.items():
    plt.plot([i/10 for i in range(101)], [sol.subs(t, t_val) for t_val in [i/10 for i in range(101)]], label=f'k={k_val}')  #Ploting
plt.xlabel('Time (t)'), plt.ylabel('y(t)'), plt.legend()  # plot labeling, legends->color bujhanor jonno
plt.show()
```

```python
import sympy as sym
import matplotlib.pyplot as plt

t = sym.symbols('t')
y = sym.Function('y')(t) #y=y(t)

k_vals = [0.1, 0.2, 0.5]
initial_condition = {y.subs(t, 0): 5}

solutions = {}
for k_val in k_vals:
    differential_equation = sym.diff(y, t) + k_val*y
    solution = sym.dsolve(differential_equation, ics=initial_condition).rhs #x+y=1 , y =5 ,x+5=1
    solutions[k_val] = solution #{x+0.1:y,0.2:x+y+5,0.3:x-y}

for k_val, solution in solutions.items():
    time_values = [i/10 for i in range(101)]
    #print(time_values)
    # for i in range(101):
    #   i=i/10
    y_values = [solution.subs(t, t_val) for t_val in time_values]
    #print(y_values)
    # for t_val in time_values:
    #   y_values= solution the equation using the value of t_val
    plt.plot(time_values, y_values, label=f'k={k_val}')

plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
